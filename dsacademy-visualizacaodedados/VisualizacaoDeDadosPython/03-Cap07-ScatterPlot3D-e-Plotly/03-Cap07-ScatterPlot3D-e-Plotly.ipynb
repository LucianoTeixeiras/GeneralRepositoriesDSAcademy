## ScatterPlot 3D e Plotly

```python
from matplotlib import pyplot
%matplotlib inline
```

```python
# Range de 0 - 541
X = range(0, 541)

# Gera valores de y com base em valores de x usando list compreheension
Y = [i*i for i in X]

# Gera o gráfico
pyplot.plot(X, Y, '-')

# Labels e título
pyplot.title('Plotting x*x')
pyplot.xlabel('Eixo X')
pyplot.ylabel('Eixo Y')

# Salva o chart como png
pyplot.savefig('Simple.png')

# Mostra o gráfico
pyplot.show()
```

```python
# Plot 3D
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.gca(projection='3d')

# Criando a curva do eixo z
x = np.linspace(0, 1, 100)
y = np.sin(x * -1 * np.pi) / 2 + 0.5

ax.plot(x, y, zs = 0, zdir = 'z', label = 'Curva randômica, apenas para o eixo z')

# Cores
colors = ('r', 'g', 'k', 'b')

for c in colors:
    x = np.random.sample(42)
    y = np.random.sample(42)
    ax.scatter(x, y, 1, zdir='x', c=c)

# Legenda
ax.legend()

# Eixos x, y e z
ax.set_xlim3d(0, 1)
ax.set_ylim3d(0, 1)
ax.set_zlim3d(0, 1)

# Mostra o chart
plt.show()
```
